## letter frequency, bigram frequency and letter proximity ###

determine frequency of individual letters and of bigrams in a corpus

also determine the conditional frequency of letters before and after a given letter, as well as co-occurrence in words

source word list is pandas dataframe with columns 'word' and 'freq'. Any other columns will be ignored.

** initial_data_munge must be run first to create pickled dataframes of word lists **

```python
dataframe_base = 'coha_words' # change as needed to point to pickle
dataframe_description = ''

data_path = 'data'
nb_path = 'letter_bigram_frequencies'

save_filename = '' #used for .pickle and .png, leave as '' to use a default filename

import pandas as pd
import os
import time

words = pd.read_pickle(data_path + '/' + dataframe_base + '.pickle')

alphabet = 'abcdefghijklmnopqrstuvwxyz'

# create lookup dicts instead of using ascii codes, because explicit is better than implicit
reverse_alpha_dict = {}
forward_alpha_dict = {}
for i in range(26):
    forward_alpha_dict[alphabet[i]] = i
    reverse_alpha_dict[i] = alphabet[i]
    
def sort_bigram(bigram):
    number_1 = forward_alpha_dict[bigram[0]]
    number_2 = forward_alpha_dict[bigram[-1]] # uses last letter, not second letter; same thing for a bigram
    if number_2 < number_1:
        return bigram[-1] + bigram[0]
    else:
        return bigram[0] + bigram[1]
    
#create empty dict with one key for every possible bigram
bigram_dict = {} # every bigram regardless of order, e.g. aa, ab, ba ...
co_dict = {} # for co-occurrence, every bigram in increasing order only, e.g. aa, ab, ac, ... bc, bd, ...
for letter_1 in alphabet:
    for letter_2 in alphabet:
        bigram_dict[letter_1+letter_2] = 0
for i in range(26):
    for j in range(26):
        if i <= j: #ever so slightly wasteful but really straightforward logic
            co_dict[alphabet[i]+alphabet[j]] = 0
```

#### create bigrams dict and dataframe

```python
redo_pickle = False #change to true to overwrite existing pickle if the underlying data changes

if save_filename == '':
    bigrams_pickle = nb_path + '/' + 'bigrams_' + dataframe_base + '.pickle'
else:
    bigrams_pickle = nb_path + '/' + save_filename + '.pickle'

if not os.path.isfile(bigrams_pickle) or redo_pickle == True:    
    
    start = time.time()
    
    for i in range(len(words)):
        wd = words.word.iloc[i]
        temp_dict = {}
        if len(wd) > 1:
            for pos in range(len(wd)):
                if pos < len(wd) - 1:
                    bigram_dict[wd[pos]+wd[pos+1]] += words.freq.iloc[i]
                    for j in range(pos, len(wd)):
                        bigram = wd[pos]+wd[j]
                        if bigram not in temp_dict.keys():
                            temp_dict[bigram] = words.freq.iloc[i]
            for key in temp_dict.keys():
                sorted_bigram = sort_bigram(key)
                co_dict[sorted_bigram] += temp_dict[key]
                
    bigrams_df = pd.DataFrame({'letter_1':['0'], 'letter_2':['0'], 'freq':[0]})
    for letter_1 in alphabet:
        for letter_2 in alphabet:
            temp = pd.DataFrame({'letter_1':[letter_1],
                                 'letter_2':[letter_2],
                                 'freq':[bigram_dict[letter_1+letter_2]]})
            bigrams_df = bigrams_df.append(temp, ignore_index=True)
    
    bigrams_df['normal_1'] = 0.0
    bigrams_df['normal_2'] = 0.0
    bigrams_df['bigram'] = ''

    for i in range(len(bigrams_df)):
        letter_1 = bigrams_df.letter_1.iloc[i]
        letter_2 = bigrams_df.letter_2.iloc[i]
        freq = bigrams_df.freq.iloc[i]
        bigrams_df.normal_1.iloc[i] = 1.0 * freq / bigrams_df[bigrams_df.letter_1 == letter_1].freq.sum()
        bigrams_df.normal_2.iloc[i] = 1.0 * freq / bigrams_df[bigrams_df.letter_2 == letter_2].freq.sum()
        bigrams_df.bigram.iloc[i] = letter_1+letter_2

    bigrams_df.set_index('bigram', drop=False, inplace=True)
    bigrams_df = bigrams_df[1:] #remove dummy row
    
    bigrams_df.to_pickle(bigrams_pickle)
    
    print "%d seconds elapsed." % (time.time() - start)
    
else:
    print 'Reading from pickle.'
    bigrams_df = pd.read_pickle(bigrams_pickle)

bigrams_df.head(10)
```

### Single letter frequencies

```python
redo_pickle = True #change to true to overwrite existing pickle if the underlying data changes

if save_filename == '':
    letfreqs_pickle = nb_path + '/' + 'letfreqs_' + dataframe_base + '.pickle'
else:
    letfreqs_pickle = nb_path + '/' + save_filename + '.pickle'

if not os.path.isfile(letfreqs_pickle) or redo_pickle == True:    
    
    start = time.time()
    
    letfreq_dict = {}
    
    for ltr in alphabet:
        letfreq_dict[ltr] = 0
    
    for i in range(len(words)):
        wd = words.word.iloc[i]
        for ltr in wd:
            letfreq_dict[ltr] += words.freq.iloc[i]
                
    letfreqs_df = pd.DataFrame()
    for letter in alphabet:
        temp = pd.DataFrame({'letter':[letter],
                              'freq':[letfreq_dict[letter]]})
        letfreqs_df = letfreqs_df.append(temp, ignore_index=True)

    letfreqs_df.to_pickle(letfreqs_pickle)
    
    letfreqs_df['normal'] = 0.0

    letfreqsum = letfreqs_df.freq.sum()

    for i in range(len(letfreqs_df)):
        letter = letfreqs_df.letter.iloc[i]
        freq = letfreqs_df.freq.iloc[i]
        letfreqs_df.normal.iloc[i] = 1.0 * freq / letfreqsum

    letfreqs_df.set_index('letter', drop=False, inplace=True)
    
    print "%d seconds elapsed." % (time.time() - start)

else:
    print 'Reading from pickle.'
    letfreqs_df = pd.read_pickle(letfreqs_pickle)
    
print letfreqs_df
```
